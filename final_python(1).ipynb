## Identify the missing data and use the appropriate method to deal with it.

```python
ko=round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#fetching the column names which are having high missing values percentage
lis2rem=[]
for i,j in ko.items():
    if j>35:
        lis2rem.append(i)
```

### 1.b. Removing columns with missing values greater than 35 %

```python
#removing columns having high percentage value of missing values
for i in df.columns:
    if i in lis2rem:
        df.drop(i,axis=1,inplace=True)
```

```python
#Appending columns with lesser missing values to a list
lis2rem=[]
for i,j in ko.items():
    if j>0 and j<35:
        lis2rem.append(i)
```

# 2. Check for outliers and removal using IQR method

```python
sns.boxplot(y=df['AMT_INCOME_TOTAL'])
plt.yscale('log')

df['AMT_INCOME_TOTAL'].describe()
```

## We will use IQR method to remove Outliers from the dataframe

```python
Q1 = df['AMT_INCOME_TOTAL'].quantile(0.25)
Q3 = df['AMT_INCOME_TOTAL'].quantile(0.75)
IQR = Q3 - Q1
df= df[~((df['AMT_INCOME_TOTAL'] < (Q1 - 1.5 * IQR)) |(df['AMT_INCOME_TOTAL'] > (Q3 + 1.5 * IQR)))]
```
